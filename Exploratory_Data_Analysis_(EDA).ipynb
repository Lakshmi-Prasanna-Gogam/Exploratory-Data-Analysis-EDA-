**Exploratory Data Analysis (EDA):**

Exploratory Data Analysis (EDA) in understanding data, identifying errors, and uncovering patterns.

**Objective**: Understand data using statistics and visualizations.

**Tools**: Pandas, Matplotlib, Seaborn, Plotly.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Data Preprocessing**

```python
dataset = pd.read_csv('/content/Titanic-Dataset.csv')
dataset
```

**Handling missing values using mean/median/dropna()/fillna()**

```python
dataset.dropna()
dataset.fillna(0)
```

**Converting categorical data into numerical data (Feature scaling)**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
catg_cols = dataset.select_dtypes(include = ['object','category']).columns
catg_cols
```

```python
dataset['Name'] = le.fit_transform(dataset['Name'])
dataset['Sex'] = le.fit_transform(dataset['Sex'])
dataset['Ticket'] = le.fit_transform(dataset['Ticket'])
dataset['Cabin'] = le.fit_transform(dataset['Cabin'])
dataset['Embarked'] = le.fit_transform(dataset['Embarked'])
```

**Generating summary statistics**

```python
dataset.describe()
```

**Histograms & Boxplots**

```python
dataset.hist(figsize=(8,9), bins=30)  #Histogram
plt.tight_layout
plt.show()
```

```python
#Boxplot  used for spotting outliers
for col in dataset.select_dtypes(include=['number']).columns:
  plt.figure(figsize=(1,1))
  sns.boxplot(dataset[col])
  plt.show()
```

**Correlation Matrix / Pairplot** used to identify relationships between features.

```python
#Correlation Matrix - Detect linear relationships
plt.figure(figsize=(8,8))
sns.heatmap(dataset.corr(), annot=True)
plt.show()
```

```python
#Pairplot - Visualize potential clusters and relationships
sns.pairplot(dataset)
plt.show()
```

**Interactive Visuals with Plotly**

```python
import plotly.express as px
```

```python
fig = px.histogram(dataset, x='Age', nbins=30, color='Survived', barmode='overlay',
                   title='Age Distribution by Survival')
fig.update_layout(width=500, height=500)
fig.show()
```

```python
fig = px.histogram(dataset, x='Fare', nbins=50, color='Pclass', barmode='overlay',
                   title='Fare Distribution by Class')
fig.update_layout(width=500, height=500)
fig.show()
```

```python
fig = px.scatter(dataset, x='Age', y='Fare', color='Survived',
                 hover_data=['Sex', 'Pclass'],
                 title='Age vs Fare by Survival')
fig.update_layout(width=500, height=500)
fig.show()
```

```python
fig = px.scatter(dataset, x='Age', y='SibSp', color='Pclass',
                 hover_data=['Survived', 'Sex'],
                 title='Age vs Siblings/Spouse Count by Class')
fig.update_layout(width=500, height=500)
fig.show()
```
